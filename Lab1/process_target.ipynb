```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df1 = pd.read_csv('averaged_times1.txt', header=None, names=['time'])
df2 = pd.read_csv('averaged_times2.txt', header=None, names=['time'])

avg1 = df1['time'].mean()
avg2 = df2['time'].mean()

print(avg1, avg2)




fig, ax = plt.subplots()
# plot

ax.bar(['a','b'], [np.mean(df1['time']), np.mean(df2['time'])])
ax.errorbar(['a','b'], [np.mean(df1['time']), np.mean(df2['time'])],yerr=[np.std(df1['time']), np.std(df2['time'])],fmt="o",color="r")
ax.set_title('Bar chart with error bars')
```

```python


fig, ax = plt.subplots()

# box plot
ax.boxplot([df1['time'], df2['time']])
ax.set_title('Box plot')

ax.set_xticks([1,2],labels=['x1', 'x2'])
ax.set_xlabel('mode')
ax.set_ylabel('time')


```

```python
# plot violin plot


fig, ax = plt.subplots()


ax.violinplot([df1['time'], df2['time']], showextrema=True)
ax.set_title('Violin plot')

ax.scatter([1,2], [np.mean(df1['time']), np.mean(df2['time'])], color='red', label='mean')
ax.set_xticks([1,2],labels=['x1', 'x2'])
ax.set_xlabel('Four separate samples')
ax.set_ylabel('Observed values')
```
